```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
#function for reuse
def encode(data):
  return LabelEncoder().fit_transform(data)
```

```python
#loading data
data = pd.read_csv('/content/drive/MyDrive/DL Projects/Classification Projects/ANN/Customer Churn Prediction/data/Churn_Modelling.csv')
data.info() #no missing values
```

```python
#removing unnecesasry data
data.drop(['RowNumber','CustomerId','Surname'], axis=1, inplace=True)
```

# Data Visualization & Feature Engineering

```python
#class balance
sns.countplot(data['Exited']) #not balanced (random sampling will be applied after splitting the dataset to prevent data leakage)
```

```python
#majority of data is from ...
sns.countplot(data['Geography'])
```

```python
#converting categorical variables
data['Gender'] = encode(data['Gender'])
data['Geography'] = encode(data['Geography'])
```

# Saving the dataset

```python
data.to_csv('/content/drive/MyDrive/DL Projects/Classification Projects/ANN/Customer Churn Prediction/data/processeddata.csv', index=False)
```
